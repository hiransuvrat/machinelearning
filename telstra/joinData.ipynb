```python
import pandas as pd

def filterCategorical(data, column, value):
    counts = dict(data[column].value_counts())
    for itr in counts:
        if counts[itr] < value:
            counts[itr] = -1
    result = []
    for index, row in data.iterrows():
        if counts[row[column]] == -1:
            result.extend(['other'])
        else:
            result.extend([row[column]])
    data[column] = result
    return data

train = pd.read_csv('./air/trainInt.csv')
#resourceType = pd.read_csv('./air/resourceFreq.csv')
resourceType = pd.read_csv('./air/resource_type.csv')
#eventType = pd.read_csv('./air/eventFreq.csv', index_col=False)
eventType = pd.read_csv('./air/eventInt.csv')
severityType = pd.read_csv('./air/severity_type.csv')
#logFeature = pd.read_csv('./air/logFreq.csv', index_col=False)
logFeature = pd.read_csv('./air/logInt.csv')
test = pd.read_csv('./air/testInt.csv')
#print logFeature

train = pd.merge(left=train, right=severityType, on='id', how='left')
train = pd.merge(left=train, right=eventType, on='id', how='left')
train = pd.merge(left=train, right=resourceType, on='id', how='left')
train = pd.merge(left=train, right=logFeature, on='id', how='left')

test = pd.merge(left=test, right=resourceType, on='id', how='left')
test = pd.merge(left=test, right=eventType, on='id', how='left')
test = pd.merge(left=test, right=severityType, on='id', how='left')
test = pd.merge(left=test, right=logFeature, on='id', how='left')

#Feature engineering
#train['i_id_locInt'] = train['id'].astype(str) + ',' + train['locInt'].astype(str)
#train['locInt'] = (train['locInt'] / 100).astype(int)
#train['id'] = (train['id'] / 1000).astype(int)
#test['i_id_locInt'] = test['feature 82'] * test['feature 203']

#test['i_id_locInt'] = test['id'].astype(str) + ',' + test['locInt'].astype(str)
#test['locInt'] = (test['locInt'] / 100).astype(int)
#test['id'] = (test['id'] / 1000).astype(int)

#train = filterCategorical(train, 'location', 0)
#test = filterCategorical(test, 'location', 0)

train.to_csv('./air/joinedTrain.csv', index=False)
test.to_csv('./air/joinedTest.csv', index=False)

train = pd.concat((train, test), axis=0, ignore_index=True)
```
